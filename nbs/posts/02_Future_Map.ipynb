```python
#| hide
import graphviz  
primary = '#FFFFFF'
secondary = '#F6AE2D'
third = '#002855'
bg = '#F1FAF0'
fifth = '#002855'

g = graphviz.Digraph('A', filename='process.gv', engine = 'dot')
g.attr('graph', style='filled', rankdir='LR', compound='true',fillcolor=bg, fontcolor=third, fontname = "Helvetica,Arial,sans-serif")
g.attr('node', style='filled', size='8,5', shape='doublecircle',fillcolor=secondary, fontcolor=third, fontname = "Helvetica,Arial,sans-serif") 
g.attr('edge', arrowhead='vee', arrowsize='1', len='1.00', fontname = "Helvetica,Arial,sans-serif")


#end goal
with g.subgraph(name='cluster_Project') as c:
    c.attr(label='Project', shape = 'doublecircle', fillcolor=bg, fontcolor=third)
    c.attr('node', shape = 'box', fillcolor=secondary, fontcolor=third)
    c.node('InfoAI', 'Information AI')
    c.node('3D', '3D vision- modeling AI')
    c.node('stocks', 'Trading Algo')
    
    
with g.subgraph(name='clusterBusiness') as v:
    v.attr(label='Business/Marketing/Finance', shape = 'doublecircle', fillcolor=bg, fontcolor=third)
    v.attr('node', shape = 'doublecircle', fillcolor=secondary, fontcolor=third)
    v.node('InfoAI', 'Information AI')
    v.node('Drone', 'Drone')
    v.node('Laser', 'Laser')

    
with g.subgraph(name='clusterIntern') as b:
    b.attr(label='Intership', shape = 'doublecircle', fillcolor=bg, fontcolor=third)
    b.attr('node', shape = 'doublecircle', fillcolor=secondary, fontcolor=third)
    b.node('phisaver', 'Phisaver')
    b.node('xplainable', 'Xplainable')
    

g.edge('FASTAI:e', 'Thesis')
g.edge('FASTAI:e', 'InfoAI:w')
g.edge('Thesis:e', '3D:w')


g.edge('phisaver:e', 'Flutter:w')
g.edge('phisaver:e', 'Forecasting:w')
g.edge('phisaver:e', 'ML')


g.edge('xplainable:e', 'ML')
g.edge('ML:e', '3D:w')

g.edge('Forecasting:e', 'stocks:w')


g.edge('Flutter:e', '3D:w')
g.edge('Flutter:e', 'Thesis:w')


g.edge('Drone:e', 'Business:w')
g.edge('Laser:e', 'Business')


g.edge('Business:e', 'InfoAI:s', lhead='cluster_Project')


g.render(format='svg', cleanup=False)

g
```
